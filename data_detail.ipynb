```python
import os
import matplotlib.pyplot as plt

# Lista para armazenar a quantidade de imagens em cada pasta
quantidades = []

# Loop através das pastas
for i in range(6):
    pasta = f"data/custom/{i}"
    # Conta o número de arquivos na pasta
    quantidade = len(os.listdir(pasta))
    quantidades.append(quantidade)

# Cria o gráfico
plt.bar(range(6), quantidades)
plt.xlabel("Pasta")
plt.ylabel("Quantidade de imagens")
plt.title("Quantidade de imagens por pasta")
plt.show()
```
